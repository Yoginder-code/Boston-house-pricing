## FDI in various sectors of Indian economy till 2021: Line chart

```python
def plot_data(data=df1,x='Year',y='Amount is million USD',t='FDI IN ALL SECTORS OF INDIAN ECONOMY'):
    plt.figure(figsize=(6,6))
    plt.plot(data)
#     plt.legend(data.columns)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(t)
    plt.show()
    
plot_data()
```

<h3>Observation</h3>
1. The line graph shows FDI in various sectors till 2021. Currently it is very difficult to identfy and extract information out of this graph so we will try to figure out the top 5 sectors where the FDI is the highest and we can look at these sectors.

## Finding out the top 5 sectors of the economy till 2021:

```python
def avg_sector(df=df,number=5):
    data = df.copy(deep=True)
    data['avg'] = pd.Series()
    for sectors in data.index:
        data['avg'] = data.mean(axis=1)
        top_df = data.sort_values(by='avg',ascending=False,inplace=False)
    top = top_df.head(number).index
    top_df.drop(columns=['avg'],axis=1,inplace=True)
    return top

 

top_sectors = avg_sector(df)
df_top = df1[top_sectors]
df_top.tail()
```

<h3> Observation </h3>
We can see that service sector,computer software,telecommunications,trading and contruction development sectors are the top 5 sectors till 2021 that is in these sectors India recieved the most of its FDI.

### Creating the dataframe for the Top 5 Sectors till 2021:

```python
Top_1 = df1["SERVICES SECTOR (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing and Analysis, Other)"]
Top_2 = df1["COMPUTER SOFTWARE & HARDWARE"]
Top_3 = df1["TELECOMMUNICATIONS"]
Top_4 = df1["TRADING"]
Top_5 = df1["CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-development projects"]

```

#### Plotting the line chart for the top 5 sectors of the economy:

```python
plt.figure(figsize=(7,7))
plt.xlabel('DATE')
# Set the y axis label of the current axis.
plt.ylabel('Amount in million USD')
# Set a title 
plt.title('FDI IN TOP 5 SECTORS OF INDIAN ECONOMY TILL 2021')
# Display the figure.
plt.plot(Top_1, color='blue',label = 'Services sector',linestyle='dashed')
plt.plot(Top_2, color='red',label = 'COMPUTER SOFTWARE & HARDWARE',linestyle='dashed')
plt.plot(Top_3, color='green',label = 'TELECOMMUNICATIONS',linestyle='dashed')
plt.plot(Top_4, color='black',label = 'TRADING',linestyle='dashed')
plt.plot(Top_5, color='orange',label = 'CONSTRUCTION DEVELOPMENT',linestyle='dashed')
# show a legend on the plot
plt.legend()
plt.show()
```

<h3> Observation</h3>
After knowing the top 5 sectors, we are using the line chart to know the trend of these sectors and we can see that red dotted line ie. computer software and hardware  is increasing and showing the upward trend which means that FDI has been increased in these years.
